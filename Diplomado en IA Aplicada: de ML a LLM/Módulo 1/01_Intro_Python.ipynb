### 🔜 **Introducción**

#### 📌 **¿Qué es la Ciencia de Datos?**

La **Ciencia de Datos** es una disciplina que combina **matemáticas**, **estadística** y **programación**. Su finalidad es generar valor a través del análisis, la predicción y la visualización de patrones ocultos en los datos.

Aplicaciones típicas incluyen:

- Predicción de enfermedades en medicina.
- Detección de fraude financiero.
- Recomendaciones en plataformas de streaming.
- Optimización de procesos industriales.

##### 🔄 Ciclo de vida de un proyecto de Ciencia de Datos

El proceso típico de trabajo en un proyecto de Ciencia de Datos incluye las siguientes etapas:

1. ✅ **Obtención de datos: recolectar información desde diferentes fuentes.**
2. ✅ **Limpieza y preparación: tratar errores, valores faltantes, formatos.**  
3. Análisis exploratorio: buscar patrones y relaciones.  
4. Modelado: aplicar técnicas estadísticas o de machine learning.  
5. Evaluación y ajuste: validar y mejorar los modelos.  
6. Despliegue: implementar el modelo para que genere valor real.

A continuación, exploraremos algunas operaciones fundamentales que se pueden realizar con strings:

- `.lower()` → Convierte todo el texto a minúsculas.
- `.upper()` → Convierte todo el texto a mayúsculas.
- `.title()` → Convierte a formato título (primera letra en mayúscula).
- `.strip()` → Elimina espacios al inicio y al final del string.
- `.replace("a", "b")` → Reemplaza una subcadena por otra.
- `.split()` → Divide el string en una lista de palabras.
- `len()` → Retorna la cantidad de caracteres en el string.
- `.count("a")` → Cuenta cuántas veces aparece una subcadena.
- `.find("x")` / `.index("x")` → Encuentra la posición de una subcadena.
- `in` → Verifica si una subcadena está contenida en otra.
- `[]` → Se usa para acceder a caracteres individuales por índice (positivos o negativos).

````python
texto = "  La Ciencia de Datos es poderosa  "
````

````python
print(texto.lower())      # '  la ciencia de datos es poderosa  '
print(texto.strip())      # 'La Ciencia de Datos es poderosa'
print(texto.title())      # 'La Ciencia De Datos Es Poderosa'
print(texto.split())      # ['La', 'Ciencia', 'de', 'Datos', 'es', 'poderosa']
print("Datos" in texto)   # True
print(texto[6:13])        # 'Ciencia'
````

##### Ejercicio 1

1. Crea un programa que le pida al usuario su nombre y su año de nacimiento. El programa debe imprimir un saludo y decirle cuántos años tiene (asumiendo que el año actual es 2026).

✅ **Ejemplo de entrada y salida esperada**

```plaintext
Nombre: Sofía  
Año de nacimiento: 2000  

Hola Sofía, tienes 26 años.

````python
nombre = input("Nombre: ")
anio_nacimiento = int(input("Año de nacimiento: "))

anio_actual = 2026
edad = anio_actual - anio_nacimiento

print(f"Hola {nombre}, tienes {edad} años.")
````

##### Ejercicio 2

2. Pide al usuario que ingrese su nombre completo (nombre y apellido) en minúsculas. El programa debe imprimir el nombre con:

- La primera letra en mayúscula (formato título)

- La cantidad de caracteres (sin contar espacios)

- Un mensaje formateado con f-string

✅ **Ejemplo de entrada y salida esperada**

```plaintext
Nombre completo: juan pérez  

Tu nombre formateado es: Juan Pérez  
Tu nombre tiene 9 letras.


````python
nombre_completo = input("Nombre completo: ")

nombre_formateado = nombre_completo.title()
cantidad_letras = len(nombre_completo.replace(" ", ""))

print(f"Tu nombre formateado es: {nombre_formateado}")
print(f"Tu nombre tiene {cantidad_letras} letras.")
````

##### Ejercicio 3

Pide al usuario que escriba una frase. El programa debe:

- Mostrar cuántas palabras tiene

- Decir cuántas veces aparece la letra "a" (minúscula)

- Mostrar la frase invertida

✅ **Ejemplo de entrada y salida esperada**

```plaintext
Frase: la ciencia de datos es fascinante  

Palabras: 6  
Número de letras 'a': 4  
Frase invertida: etnancisaf se sotad ed aicneic al

````python
frase = input("Frase: ")

num_palabras = len(frase.split())
num_a = frase.count('a')
frase_invertida = frase[::-1]

print(f"\nPalabras: {num_palabras}")
print(f"Número de letras 'a': {num_a}")
print(f"Frase invertida: {frase_invertida}")
````
