#### 📋 **Listas (`list`)**

````python
# Definición: colección ordenada y mutable
lista = [10, 20, 30, 40, 50]
````

````python
# Indexación y slicing
primer_elemento = lista[0]
ultimos_dos = lista[-2:]
````

````python
# Métodos comunes
lista.append(60)           # Agrega un elemento al final
lista.insert(2, 25)        # Inserta 25 en la posición 2
lista.remove(40)           # Elimina el primer 40
valor_eliminado = lista.pop()  # Elimina y devuelve el último valor
lista.sort()               # Ordena la lista en su lugar
lista.reverse()            # Invierte el orden de la lista
````

````python
# Iteración
for elemento in lista:
    print(f"Elemento: {elemento}")
````

````python
# List comprehension
cuadrados = [x**2 for x in range(6)]
cuadrados
````

#### 🔐 **Tuplas (`tuple`)**

````python
# Definición: colección ordenada e inmutable
tupla = (1, 2, 3, 4)
````

````python
# Indexación
segundo = tupla[1]
````

````python
# Desempaquetado
a, b, c, d = tupla
````

````python
# Tuplas anidadas
coordenadas = ((1, 2), (3, 4))
````

#### 🗂️ **Diccionarios (`dict`)**

````python
# Definición: colección no ordenada de pares clave-valor, mutable
persona = {
    "nombre": "Ana",
    "edad": 28,
    "ciudad": "Bogotá"
}
````

````python
# Acceso
nombre = persona["nombre"]
edad = persona.get("edad")  # Mejor para evitar errores si no existe la clave
````

````python
# Métodos comunes
persona["profesión"] = "Ingeniera"  # Añadir clave
persona.update({"edad": 29})        # Actualizar clave existente
profesion = persona.pop("profesión") # Eliminar y devolver valor
````

````python
# Claves, valores, items
claves = list(persona.keys())
valores = list(persona.values())
items = list(persona.items())
````

````python
# Iteración
for clave, valor in persona.items():
    print(f"{clave}: {valor}")
````

````python
# Diccionario por comprensión
cuadrados_dict = {x: x**2 for x in range(5)}
cuadrados_dict
````

#### ♻️ **Conjuntos (`set`)**

````python
# Definición: colección no ordenada de elementos únicos
conjunto = {1, 2, 3, 4, 5}
conjunto2 = {4, 5, 6, 7}
````

````python
# Métodos
conjunto.add(6)
conjunto.remove(1)
````

````python
# Operaciones de conjuntos
union = conjunto | conjunto2
interseccion = conjunto & conjunto2
diferencia = conjunto - conjunto2
diferencia_simetrica = conjunto ^ conjunto2
````

````python
# Iteración
for elem in conjunto:
    print(f"Elemento único: {elem}")
````

#### 📊 **Dataframe (`pandas.DataFrame`)**

##### Ejercicio 1

1. Pide al usuario ingresar las edades de un grupo de personas (hasta que escriba "fin"). Luego:

- Muestra la cantidad total de edades ingresadas.

- Muestra la lista de edades únicas (sin repetir) usando un conjunto.

- Calcula el promedio de edad.

✅ **Ejemplo de entrada y salida esperada**

```plaintext
Ingresa una edad (o "fin" para terminar): 25  
Ingresa una edad (o "fin" para terminar): 30  
Ingresa una edad (o "fin" para terminar): 25  
Ingresa una edad (o "fin" para terminar): fin  

Total de edades: 3  
Edades únicas: [25, 30]  
Promedio de edad: 26.7

````python
edades = []

while True:
    entrada = input("Ingresa una edad (o \"fin\" para terminar): ")
    if entrada.lower() == "fin":
        break
    try:
        edad = int(entrada)
        edades.append(edad)
    except ValueError:
        print("Edad inválida.")

if edades:
    unicas = list(set(edades))
    promedio = sum(edades) / len(edades)

    print(f"\nTotal de edades: {len(edades)}")
    print(f"Edades únicas: {unicas}")
    print(f"Promedio de edad: {round(promedio, 1)}")
````

##### Ejercicio 2

1. Pide al usuario ingresar nombres de estudiantes y sus notas (hasta escribir "fin"). Luego:

- Crea un diccionario con los nombres como claves y las notas como valores.

- Convierte el diccionario en un DataFrame.

- Muestra el promedio general.

- Filtra a los estudiantes que tienen nota mayor o igual a 3.0.

**✅ Ejemplo de entrada y salida esperada**

```plaintext
Nombre: Ana  
Nota: 4.5  
Nombre: Luis  
Nota: 2.8  
Nombre: fin  

Promedio general: 3.65  
Estudiantes aprobados:
    Nombre  Nota
0     Ana   4.5

````python
import pandas as pd

notas = {}

while True:
    nombre = input("Nombre: ")
    if nombre.lower() == "fin":
        break
    try:
        nota = float(input("Nota: "))
        notas[nombre] = nota
    except ValueError:
        print("Nota inválida.")

df = pd.DataFrame(list(notas.items()), columns=["Nombre", "Nota"])

promedio = df["Nota"].mean()
aprobados = df[df["Nota"] >= 3.0]

print(f"\nPromedio general: {round(promedio, 2)}")
print("Estudiantes aprobados:")
print(aprobados)
````
