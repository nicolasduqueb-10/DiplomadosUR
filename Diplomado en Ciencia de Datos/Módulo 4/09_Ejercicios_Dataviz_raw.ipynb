### 📋 Descripción de las variables del dataset `penguins`

Este dataset proviene del estudio *Palmer Station LTER* en la Antártida. Cada fila representa un pingüino individual observado.

| **Columna**            | **Tipo**     | **Descripción**                                                                 |
|------------------------|--------------|----------------------------------------------------------------------------------|
| `species`              | categórica   | Especie del pingüino (`Adelie`, `Chinstrap`, `Gentoo`).                         |
| `island`               | categórica   | Isla donde fue observado el pingüino (`Biscoe`, `Dream`, `Torgersen`).          |
| `bill_length_mm`       | numérica     | Longitud del pico (en milímetros).                                              |
| `bill_depth_mm`        | numérica     | Profundidad del pico (en milímetros).                                           |
| `flipper_length_mm`    | numérica     | Longitud de la aleta (en milímetros).                                           |
| `body_mass_g`          | numérica     | Masa corporal del pingüino (en gramos).                                         |
| `sex`                  | categórica   | Sexo del pingüino (`Male`, `Female`). Puede contener valores nulos.            |
| `year`                 | numérica     | Año en que se tomaron los datos (2007, 2008 o 2009).                            |

> 💡 **Nota**: este dataset contiene algunos valores faltantes, especialmente en `sex` y medidas físicas. Puedes usar `df.isna().sum()` para identificarlos.
