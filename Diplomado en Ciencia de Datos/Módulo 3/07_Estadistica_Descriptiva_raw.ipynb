### 📊 **Definición de Covarianza**

$$ COV(X, Y) = \frac{\sum_{i=1}^{n} (x_i - \mu_X)(y_i - \mu_Y)}{n}$$

* $n := $ Cantidad de datos
* $\mu_X :=$ Media de la variable $X$
* $\mu_Y :=$ Media de la variable $Y$

```python
X = 'age'
Y = 'hours_of_sleep_per_day'
n = len(df)

((df[X] - df[X].mean()) * (df[Y] - df[Y].mean())).sum()/(n-1)
```

```python
# Covarianza
df[[X, Y]].cov()
```
