# 🦖 **Data Mining con PokeAPI**

### 🔜 **Introducción**

##### **¿Qué es Data Mining y por qué es tan importante en Ciencia de Datos?**

El **Data Mining** es el proceso de extraer información útil y patrones relevantes a partir de grandes volúmenes de datos.

En Ciencia de Datos, los datos rara vez vienen listos para analizar. Generalmente:
- No tienen formato tabular
- Contienen estructuras complejas
- Necesitan limpieza y transformación

El proceso típico es:

> API → JSON → Exploración → Transformación → DataFrame

##### **¿Qué es un API?**

Un **API (Application Programming Interface)** permite que diferentes aplicaciones se comuniquen entre sí.

En Ciencia de Datos los APIs son fundamentales porque:
- Permiten acceso automatizado a información
- Devuelven datos en formato JSON
- Son ampliamente utilizados en servicios web

Trabajaremos con la **PokeAPI**:
https://pokeapi.co/

### 🧙 **Manos a la obra**

##### **Paso 1: Conectarnos al API**

> Endpoint de ejemplo:
> https://pokeapi.co/api/v2/pokemon/pikachu

```python
import requests
```

```python
url = "https://pokeapi.co/api/v2/pokemon/pikachu"
response = requests.get(url)

response.status_code
```

```python
data = response.json()

type(data)
```

```python
data
```

##### **Paso 2: Explorando la estructura del JSON**

```python
data.keys()
```

```python

data["name"], data["height"], data["weight"]
```

##### **Paso 3: Navegando estructuras anidadas**

```python
data["types"]
```

```python
data["types"][0]["type"]["name"]
```

##### **Paso 4: Estructurando la información básica**

```python
pokemon_info = {
    "name": data["name"],
    "height": data["height"],
    "weight": data["weight"],
    "base_experience": data["base_experience"]
}

pokemon_info
```

```python

import pandas as pd

df_pokemon = pd.DataFrame([pokemon_info])
df_pokemon
```

##### **Paso 5: Extrayendo estadísticas (stats)**

```python
data["stats"]
```

```python
stats_list = []

for s in data["stats"]:
    stats_list.append({
        "stat": s["stat"]["name"],
        "base_stat": s["base_stat"]
    })

stats_list
```

```python
df_stats = pd.DataFrame(stats_list)
df_stats
```

##### **Paso 6: Feature Engineering**

```python
power_total = df_stats["base_stat"].sum()
power_total
```

```python
df_pokemon["power_total"] = power_total
df_pokemon
```
